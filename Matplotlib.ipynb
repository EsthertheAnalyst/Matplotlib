```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
a = np.arange(1,8)
b= a**2
```

```python
plt.plot(a,b)
```

```python
plt.plot(a,b,'r')
```

```python
plt.plot(a,b,'y')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X against Y')
```

```python
plt.subplot(1,2,1)
plt.plot(a,b,'y')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X against Y')
plt.subplot(1,2,2)
plt.plot(b,a,'g')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X against Y')
```

```python
plt.subplot(2,1,1)
plt.plot(a,b,'y')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X against Y')
plt.subplot(2,1,2)
plt.plot(b,a,'g')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X against Y')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.6])
axes.plot(a,b,'y-.')
axes.set_xlabel('X1')
axes.set_ylabel('Y1')
axes.set_title('X1 against Y1')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.6,0.4,0.3])#left,bottom,width,height
axes1.plot(a,b,'y-.')
axes1.set_xlabel('X1')
axes1.set_ylabel('Y1')
axes1.set_title('X1 against Y1')
axes2 = fig.add_axes([0.1,0.1,0.4,0.3])
axes2.plot(b,a,'g-.')
axes2.set_xlabel('X2')
axes2.set_ylabel('Y2')
axes2.set_title('X2 against Y2')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.6,0.6,0.8])#left,bottom,width,height
axes1.plot(a,b,'y-.',label='a,b')
axes1.plot(b,a,'g-.',label='b,a')
axes1.set_xlabel('X1')
axes1.set_ylabel('Y1')
axes1.set_title('X1 against Y1')
axes1.legend(loc=4)
```

```python
fig,axes=plt.subplots()
axes.plot(a,b,'y-.',label='a,b')
axes.plot(b,a,'g-.',label='b,a')
axes.set_xlabel('X1')
axes.set_ylabel('Y1')
axes.set_title('X1 against Y1')
axes.legend(loc=4)
```

```python
fig,axes=plt.subplots(2,1)
axes[0].plot(a,b,'y-.',label='a,b')
axes[0].plot(b,a,'g-.',label='b,a')
axes[0].set_xlabel('X1')
axes[0].set_ylabel('Y1')
axes[0].set_title('X1 against Y1')
axes[0].legend(loc=4)
axes[1].plot(a,b,'g-.',label='a,b')
axes[1].plot(b,a,'r-.',label='b,a')
axes[1].set_xlabel('X1')
axes[1].set_ylabel('Y1')
axes[1].set_title('X1 against Y1')
axes[1].legend(loc=2)
```
